**Answer :**بهترین تکنیک عادی سازی تکنیکی است که از نظر تجربی به خوبی کار می کند، بنابراین اگردر توزیع ویژگی ما به خوبی کار می کنند، ایده های جدیدی را باید امتحان کنیم. هنگامی که ویژگی کم و بیش به طور یکنواخت در یک محدوده ثابت توزیع شده است. زمانی که این ویژگی حاوی برخی نقاط پرت شدید باشد
عادی سازی داده ها برای اطمینان از اینکه اندازه گیری فاصله وزن برابری برای هر متغیر دارد مهم است. بدون نرمال‌سازی، متغیری که بیشترین مقیاس را دارد بر اندازه‌گیری غالب خواهد بود. پس خروجی های مرتبط در مقیاس اصلی و غیر عادی گزارش می شوند.


```python
np.random.seed(100)
X, _ = make_blobs(n_samples=50, centers=3, n_features=1)
df = pd.DataFrame(X, columns=['1'])


kmeans = KMeans(n_clusters=5)
y = kmeans.fit_predict(df[['1']])


df['Cluster'] = y

print(df.head())


#  normalize 
X_norm = preprocessing.normalize(X)
df = pd.DataFrame(X_norm, columns=['1'])
y = kmeans.fit_predict(df[['1']])
df['Cluster'] = y
print(df.head())
```
